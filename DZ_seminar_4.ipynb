Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.info()
```

```python
df_sqft = df.set_index('price')['sqft_living'].reset_index()
plt.scatter(df_sqft['sqft_living'],df_sqft['price'])
plt.title('Зависимость стоимости недвижимости от ее жилой площади')
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость');
```

По данной диаграмме рассеивания мы видим, что между стоимостью недвижимости и ее жилой площадью имеется прямая линейная зависимость. Чем больше площадь, тем выше стоимость недвижимости.

```python
df_br = df.pivot_table(index = 'bedrooms',
                      values = 'price',
                      aggfunc = 'mean')
df_br
```

```python
plt.plot(df_br.index, df_br.values)
plt.title('Распределение средней стоимости недвижимости в зависимости от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Средняя стоимость');
```

По графику мы видим, что нет прямой зависимость стоимости недвижимости от количества спален.

```python
df_cond = df.pivot_table(index = 'condition',
                      values = 'price',
                      aggfunc = 'mean').reset_index()
df_cond.head()
```

```python
plt.bar(df_cond['condition'], df_cond['price'])
plt.title('Распределение стоимости недвижимости в зависимости от состояния')
plt.xlabel('Состояние')
plt.ylabel('Средняя стоимость');
```

По полученной гистрограмме видно, что чем лучше состояние недвижимости, тем выше ее средняя цена.

```python
df_yr = df.set_index('price')['yr_built'].reset_index()
plt.scatter(df_yr['yr_built'],df_yr['price'])
plt.title('Зависимость стоимости недвижимости от года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость');
```

Из данной диаграммы рассеивания видим, что стоимость распределена равномерно не зависимо от года постройки.
